# Data Processing and Creating a Model

```python
X = df.drop('price', axis=1)
y = df['price'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
X_train.shape
```

```python
#Looks like we have 19 incoming features; we can have 19 neurons in our model
```

```python
model = Sequential()
```

```python
model.add(Dense(19, activation='relu'))
model.add(Dense(19, activation='relu'))
model.add(Dense(19, activation='relu'))
model.add(Dense(19, activation='relu'))

model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')
```

```python
#we are training with X_train and Y_train as we are going we are testing with our test data
model.fit(x=X_train, y= y_train, validation_data=(X_test, y_test), batch_size=128, epochs=400)
```

# Model Evaluation and Predictions

```python
# we can get histrory of those losses
model.history.history
```

```python
#  since we requested validation it will provide the validation losses too; in the way we can compare the loss on training set vs test set;
# In order to see if i'm overfitting the model
losses = pd.DataFrame(model.history.history)
```

```python
losses.plot()
```

```python
#There is decrease in training loss and validation loss and are aligned 
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, explained_variance_score
```

```python
predictions = model.predict(X_test)
```

```python
np.sqrt(mean_squared_error(y_test, predictions))
```

```python
mean_absolute_error(y_test, predictions)
```

```python
df['price'].describe()
```

```python
5.402966e+05
```

```python
# The average house price is around 540296.6 and our mean absolute error is around 100k ; This is not really great, we are off by 100k ;
```

```python
#Best possible score is 1.0, lower values are worse.
explained_variance_score(y_test, predictions)
```

```python
plt.scatter(y_test, predictions)
plt.plot(y_test, y_test, 'r')
```

```python
#In the above the red line represent the best or prefect predicition line. we are good at predicting house between 0.3 and 2 million dollars
```
