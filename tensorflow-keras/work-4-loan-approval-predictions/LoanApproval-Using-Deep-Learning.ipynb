# Preprocessing 

```python
from sklearn.model_selection import train_test_split
```

```python
df['loan_repaid']
```

```python
df = df.drop('loan_repaid', axis=1)
```

```python
df.corr()
```

```python
X = df.drop('loan_status', axis=1).values
```

```python
y= df['loan_status']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

# Create and training the model

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
```

```python
#Lets start of Dense layer with our train data shape
```

```python
X_train.shape
```

```python
model = Sequential()
model.add(Dense(77, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(36, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(19, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(units=1, activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam')
```

```python
model.fit(x=X_train, y=y_train, epochs=25, batch_size=256, validation_data=(X_test, y_test))
```

```python
#Let's save our model
```

```python
from tensorflow.keras.models import load_model
```

```python
model.save('LoanApprovalModel')
```

```python
losses = pd.DataFrame(model.history.history)
```

```python
losses.plot()
```

```python
# Create predictions from the X_test set and display classification report and confusion matrix for the X_test set
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
predictions = model.predict(X_test)
# Ensure that target values are integers
y_test = y_test.astype(int)
```

```python
print(classification_report(y_test, predicted_classes))
```
